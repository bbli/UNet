```python
import pandas as pd
```

```python
from skimage import io
```

```python
label=io.imread('train-labels.tif')
```

```python
label.shape
```

```python
pic= label[0,:,:]
def row(index):
    return pic[index,:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(index):
    plt.plot(row(index),linestyle='None',marker='o')
```

```python
plot(0)
```

```python
plot(256)
```

```python
plot(322)
```
